```python
import numpy as np 
import pandas as pd
np.random.seed(1)
```

```python
def random_sequence_generator():
    
    sequence=[]
    final_sequence=[]

    for count in range(900):
        
        number_sequence=np.random.randint(1,46,7)


        if sum(number_sequence)<=180 and sum(number_sequence)>=120:
            if np.count_nonzero(number_sequence%2==0)==4 and np.count_nonzero(number_sequence%2!=0)==3 or np.count_nonzero(number_sequence%2==0)==3 and np.count_nonzero(number_sequence%2!=0)==4:
                
                sequence.append(number_sequence)
            
            else:
                pass

        else:
            pass

    for i in sequence:

        
        unique_sequence=[]
        
        for j in i:
            
            if j not in unique_sequence:
                unique_sequence.append(j)
            
        if len(unique_sequence)==7:
            final_sequence.append(unique_sequence)

    return final_sequence
```

```python
Lottery_Lots=random_sequence_generator()
sequence_pool=pd.DataFrame(Lottery_Lots,
                columns=[['Number-1','Number-2','Number-3','Number-4','Number-5',
                'Number-6','Number-7']])

sequence_pool['Sum']=sequence_pool.sum(axis=1)
```
